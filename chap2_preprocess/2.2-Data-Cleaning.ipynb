### 数据错误
- Outliers：离群值
- Rule violations：违反非空、唯一等规则
- Pattern violations：违反语义或格式规范，如错拼、单位不一致、空格等

处理方法：

- 对类别：去掉长尾类，或者直到如何修复的情况下手动 fix + merge
- 对数值：箱线图box height(25%-75%的差值)，箱子边缘上下 1.5 倍 box height 规定为上下界，丢弃超出范围的 outliers


基于规则的检测：

- Functional dependencies：x $\rightarrow$ y means 同一个x对应唯一一个y(如邮政编码决定所在州)，若不符合，直接删除或手动修正
- Denial constraints：一些更灵活的一阶逻辑表达式。

基于模式的检测：
- 语法检测 Syntactic patterns：
    1. 统一样式：eng, english $\rightarrow$ English
    2. 统一类型：一列绝大部分是整数型，设置为整数型，手动修正或丢弃非整数值
- 语义检测 Semantic patterns：基于知识图谱的规则，如国家的首都
