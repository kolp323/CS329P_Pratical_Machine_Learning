## Image Transformations

Downsampling and cropping:
- 减少存储、传输成本；加快训练
- jpeg 图片质量，中等(80%-90%)时会造成 1% 精度的下降

Image whitening: PCA
- **降低特征之间的相关性**，让每一个特征（像素或分量）都能提供独立的信息
- 找到变换矩阵W，使得变换 $y = Wx$ 后的y的协方差矩阵变成单位矩阵
- 当输入数据被“白化”后，损失函数的表面会变得更像一个圆形的碗，而不是一个扁长的山谷，可以加快模型收敛

## Video Transformation

平衡：存储、质量、读取速度。

通常使用很短的片段训练：< 10s

通常是采样视频中的帧进行训练：
- 不同的视频格式和压缩质量有不同的压缩算法，相应的读取和采样(计算)成本就会加大
- 若先采样得到图片存储下来，则增加存储成本
- 一般采用 GPU 来解码和采样，计算成本小于存储成本
- 可以采取图片变换手段处理

## Text Transformation

- **Stemming and lemmatization** 词根化/语法化: a word $ \rightarrow $ a common base form 。主要目的是理解词义而非语法正确性
  - E.g. am, are, is $\rightarrow$ be; car, cars, car's, cars' $\rightarrow$ car  
  - Example: Topic modeling  
- **Tokenization** 词元化: text string $\rightarrow$ a list of tokens (smallest unit to ML algorithms)  
  - By word: `text.split(' ')`  
  - By char: `text.split('')`  
  - By subwords: 出现字典里没有的词时，切分为子词
    - e.g. "a new gpu!"$\rightarrow$ "a", "new", "gp", "##u", "!"  
    - Custom vocabulary learned from the text corpus (Unigram, WordPiece)
