### 神经网络的（架构）超参数

- Topological structure: resnet-ish, mobilenet-ish, \#layers  
- Individual layers: 
    - kernel\_size
    - channels in convolutional layer
    - hidden\_outputs in dense/recurrent layers

### NAS with Reinforcement Learning
- 通过 RNN 基于提示和评估反馈 不断生成新的模型架构
- 缺点：价格非常高昂，需要长时间大量学习才能得到比较好的结果

### The One-shot Approach
- 同时预测模型架构和训练超参数
- 构建并训练一个包含了许多架构模型的大模型
- 使用近似指标，只关注候选架构之间的排序（类似于multi-fidelity）
- 重新从头训练预测最好的 架构

**可微架构搜索，Differentiable Architecture Search**
- 拓扑结构先确定下来，选择每个层的超参数

- Relax the categorical choice to a softmax over possible operations:  
  - 每层有多个候选  
  - Output of $i$-th candidate at layer $l$is$o_i^l$
  - Select操作子处可学习权重 $\mathbf{a}^l$. $(i+1)$-th layer 的输入是 $\sum_i \alpha_i^l o_i^l$ with $\boldsymbol{\alpha}^l = \text{softmax}(\mathbf{a}^l)$
  - Choose candidate $\arg\max_i \alpha_i$
  - Jointly learn $\mathbf{a}^l$ and network parameters  
- 给softmax操作加上温度，使 $\mathbf{a}^l$ 更趋向于0或1而非中间值：A more sophisticated version (DARTS) achieves SOTA and reduces the search time to ~3 GPU days

### Scaling CNNs  

- A CNN can be scaled by 3 ways:  
  - Deeper: more layers  
  - Wider: more output channels  
  - Larger inputs: increase input image resolutions  
- EfficientNet proposes a compound scaling:  
  - Scale depth by $\alpha^\phi$, width by $\beta^\phi$, resolution by $\gamma^\phi$（乘原来的数来扩大）
  - $\alpha \beta^2 \gamma^2 \approx 2$(so FLOP increases by 2× when $\phi = 1$)  
  - Tune $\alpha, \beta, \gamma, \phi$

## Research directions  

- Explainability of NAS result  
- Search architecture to fit into edge devices  
  - Edge devices are more and more powerful, data privacy concerns

  - Minimize both model loss and hardware latency
