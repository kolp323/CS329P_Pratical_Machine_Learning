# 3.2 决策树

**优点**：
- 可解释：常用于银行业、保险业
- 可处理数值和类别特征
- 简单、无需太多调参、效果不错，一般可用时是首选（或试探模型）

**缺点**：
- 非常不稳定，易受噪音影响（可通过集成学习帮助解决）
- 复杂的树导致过拟合（通过剪枝prune trees解决）
- 难以并行化

### 随机森林 Random Forest
- 训练多个决策树来提高稳定性
- 每个树独立训练
- 类别：投票   数值：平均值

**随机性的来源**：
- Bagging：随机可重复采样（行随机采样）
- 随机选择特征的子集（列随机采样）

### 梯度提升决策树 Gradient Boosting Decision Trees
- 顺序化地训练多个决策树
- 后一个树预测前面所有树预测值之和 与 真实值 之间的残差


- $F_t(x)$表示之前所有树的和，训练的新树 $f_t$ 预测$\{(x_i, y_i - F_t(x_i))\}_{i=1,...}$
- $F_{t+1}(x) = F_t(x) + f_t(x)$
- 残差等价于 $-\partial L/ \partial F$，若使用均方误差。即每次训练新的树来拟合梯度的负数，即梯度下降
