# 3.1 机器学习模型介绍

### ML 算法类型
- 监督学习：自监督学习（标号来自数据本身）
- 半监督学习
- 无监督学习：无标注数据，如聚类、density estimation（GAN生成对抗模型）
- 强化学习：模型与环境交互，获取观察点，机器做出行为，得到环境反馈

### 监督学习的组件
- 模型
- Loss：衡量预测值与真实值之间的差异
- Objective 目标函数
- Optimization 优化算法

### 监督学习模型类型
- 决策树 Decision trees
- Linear methods
- 核方法 Kernel machines：使用核函数衡量两个样本之间的相似度
- Neural Networks
