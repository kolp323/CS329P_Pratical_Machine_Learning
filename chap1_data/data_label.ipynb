## 半监督学习SSL

三个假设：
- **Continuity assumption**: 相似特征的数据往往有相同标号
- **Cluster assumption**: 数据有聚类倾向，处于一类的数据往往有相同标号
- **Manifold assumption**: 数据本质上所分布的流形的维度低于输入维度

自训练过程：  
只保留置信度高的伪样本

## 众包标注

任务简化：多分类化为多个二分类任务

任务量减少：Active Learning
- 人工干预：选择最感兴趣的类别进行标注
- Uncertainnity sampling：选择最不确定类别的样本进行标注
- Query-by-committee：类似于自训练，训练多个模型进行重要度投票

<img src="paste_image/2026-01-27-20-38-23.png" width="50%">

质量控制：
- 重复标注投票：最简单且效果最好
- 只重复发送有冲突标注的样本：初始两次，不一致时再多法几次
- 筛选标注工：使用已标注样本检测标注工准确率

## 弱监督学习

半自动地生成标号，比人工效果稍差。

数据编程：总结部分规律，用程序启发式地赋予标号
