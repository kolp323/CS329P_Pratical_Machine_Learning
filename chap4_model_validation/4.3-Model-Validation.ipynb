# 4.3 模型验证

### 估计泛化误差

test dataset: 只能使用一次

validation dataset: 可以被使用多次

### 获得验证集

将数据集按一定比例划分为 训练集 和 验证集。

- 随机划分：样本满足 IID 独立同分布时
- 随时间连续变化的数据：(房价、股票)验证集应该在训练集的时间之后
- 多个样本属于一组：按组随机划分(同一个人的人脸图片，要么在训练集，要么在验证集)
- 不平衡数据：多采样类少样本

### K-fold Cross Validation
数据较少时使用

1. 将数据分为 k 部分
2. 重复 k 次：每次使用第 i 部分作为验证集，其余作为训练集。
3. 将 k 次验证误差的平均值报告为验证误差。

通常 k = 5 或 10

### 常见错误
- 数据冗余：导致训练集和验证集出现了相同的样本
- 信息泄露：时序性预测时透露了未来信息、先合成再划分
