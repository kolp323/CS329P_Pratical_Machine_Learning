# 4.1 模型评估指标

### Metrics for Classification  

- **Accuracy**: $\frac{\text{\# correct predictions}}{\text{\# examples}}$

  `sum(y == y_hat) / y.size`  


- **Precision**: $\frac{\text{\# True positive}}{\text{\# (True positive + False positive)}}$

  `sum((y_hat == 1) & (y == 1)) / sum(y_hat == 1)`  

  关注 Precision 时，模型只会在置信度非常高时才预测为正类。


- **Recall**: $\frac{\text{\# True positive}}{\text{\# Positive examples}}$

  `sum((y_hat == 1) & (y == 1)) / sum(y == 1)`  

  关注 Recall 时，模型会偏向于预测为正类。

- Be careful of division by 0  

- **F1**: 平衡精度和召回，the harmonic mean of precision and recall:$\frac{2pr}{p + r}$

### AUC-ROC
选择一个阈值 $\theta$，只有当 $\hat{y} > \theta$ 时才预测为正类

ROC曲线的每一个点都代表一个 $\theta$ 值，横轴 FPR 表示负样本的错误率，纵轴 TPR 表示正样本的正确率。

AUC 是 ROC 曲线的面积，范围在 [0.5, 1]（[0, 0.5] 只需要反转类别即可）

### Business Metrics for Displaying Ads
同时优化收入和用户体验。模型预测每次是否点击，进而得到 CTR。

- Latency: 展示延迟
- ASN: average \#ads shown in a page，平均广告数量 
- CTR: actual user click through rate，真实的上线点击率
- ACP: average price advertiser pays per click，每次点击价格
- revenue $= \text{\#pageviews} \times \text{ASN} \times \text{CTR} \times \text{ACP}$


Model $\rightarrow$ Business Metrics
- The key model metric is AUC  
- A new model with increased AUC may harm business metrics, possible reasons:  
  - Lower estimated CTR $\rightarrow$ less ads displayed  
  （比如做调整后所有广告预测的CTR都降低了0.1，但是区分度/AUC不降，由于仍然按照 CTR > 0.9 的阈值进行上线展示筛选，导致展示的广告数变少）
  - Lower real CTR because we trained and evaluated on past data  
  （用旧数据预测新广告的 CTR ，实际可能偏低）
  - Lower prices  
- Online experiment: deploy models to evaluate on real traffic data
