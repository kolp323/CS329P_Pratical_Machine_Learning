### 模型复杂性
主要因素：
- 可学习的参数数量
- 参数的取值范围

最精确的衡量标准：VC dimension
- 模型能够 shatter（完全打散/拟合） 的最大样本数量

决策树的复杂度：层数的指数级

### 数据复杂性

主要因素：
- 样本数
- 每个样本的特征数
- 时间/空间结构
- 多样性
