# 5.1 偏差和方差

### 模型方差与偏差的概念
- Sample data $D = \{(x_1, y_1), \dots, (x_n, y_n)\}$ from $y = f(x) + \varepsilon$
- Learn $\hat{f}_D$ from data $D$ by minimizing MSE:$\min_{\hat{f}_D} \sum_{(x_i, y_i) \in D} \left( y_i - \hat{f}_D(x_i) \right)^2$
- We want $\hat{f}_D$ generalizes well to an unseen data point $(x, y)$ over different choices of $D$ .

### 泛化误差的组成
$泛化误差 = 模型偏差^2 + 模型方差 + 噪声方差$

$$\begin{aligned}
\mathrm{E}_D \left[ \left( y - \hat{f}_D(x) \right)^2 \right] &= \mathrm{E}_D \left[ \left( \left( f - \mathrm{E}_D[\hat{f}_D] \right) + \varepsilon - \left( \hat{f}_D - \mathrm{E}_D[\hat{f}_D] \right) \right)^2 \right] \\
&= \left( f - \mathrm{E}_D[\hat{f}_D] \right)^2 + \mathrm{E}_D \left[ \left( \hat{f}_D - \mathrm{E}_D[\hat{f}_D] \right)^2 \right] + \mathrm{E}[\varepsilon^2] \\
&= \operatorname{Bias}[\hat{f}_D]^2 + \operatorname{Var}[\hat{f}_D] + \sigma^2
\end{aligned}$$

### 减少模型偏差和方差的方法 （最终未来减小泛化误差）

**减少偏差**：
- 用更复杂的模型
- Boosting
- Stacking

**减少方差**：
- 用更简单的模型
- 正则化（L1、L2正则化）
- Bagging
- Stacking

**减少噪声误差$\sigma^2$**：
- 改进数据
