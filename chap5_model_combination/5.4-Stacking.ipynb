# 5.4 Stacking

- 将多个**不同种类模型**的输出通过全连接层做融合
- 每个模型使用完整的训练集进行独立训练（不需要bootstrap引入多样性）
- 主要是降低模型方差（多层stacking可降低偏差）

- 如果一个模型加入后不能提升精度，可以放弃
- 如果一个模型加入后带来过高推理代价，可以放弃

### Multi-layer Stacking
- 降低偏差
- 下一层的输入也会 concat 上一层的输入
- 层内模型训练是并行的
- 极易过拟合

<img src="paste_image/2026-01-29-17-24-44.png" width="58%">

### 避免过拟合的方法
- 不同层使用不同数据训练：
    - 将数据划分为 A 和 B，第一层用 A 训练
    - 将 B 经过第一次推理的结果用于第二层训练
- Repeated k-fold bagging: k*l个小模型
    1. Train k models as in k-fold cross validation
    2. 每折在验证集上的推理结果合并起来，作为下一层的训练数据
    3. （更昂贵的做法，进一步降低方差）对于每一层的模型，重复1、2操作 n 次，将 n 次输出取平均作为下一层训练数据 （k*n*l）
