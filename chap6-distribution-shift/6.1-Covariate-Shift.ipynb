# 6.1 分布偏移

### 训练与测试的不一致
- 泛化性能差：训练分布(Empirical)和测试分布不一致 $p_{emp}(x,y) \neq p(x,y)$ 。
- 协变量偏移：特征的分布变了 $p(x) \neq q(x)$，但分类规则 $p(y|x)$ 没变
- 标签偏移：结果的比例变了 $p(y) \neq q(y)$

### 协变量偏移 (Covariate Shift)
特征的分布（Covariate distribution）发生了变化，但分类或回归的规则（Label dependency）保持不变。
如：语音识别训练数据为北方口音，应用于南方口音；地理差异、光照差异等


**协变量偏移的数学修正：密度比（Density Ratio）**

定义密度比为：测试集分布 / 训练集分布

$$\alpha(x) = \frac{q(x)}{p(x)}$$

$q(x)$ 是测试集的分布，$p(x)$ 是训练集的分布

**估算密度比的方法**
- 给训练集样本贴上标签 $y=1$，给测试集（或无标签的线上数据）贴上标签 $y=-1$
- 训练一个分类器（比如逻辑回归）来预测一个样本 $x$ 到底属于哪一组 。
- 分类器会输出样本属于测试集的概率 $r(y=-1|x)$ 和属于训练集的概率 $r(y=1|x)$
$$\alpha(x) = \frac{q(x)}{p(x)} ≈ \frac{r(y=-1|x)}{r(y=1|x)}$$


### **重加权策略**
- 在训练时，给**训练集少见但测试集常见**的样本更大的权重。（目标函数改为按密度比加权平均）
- 不足：遇到训练集中罕见但测试集中偶尔出现的样本，$\alpha(x)$ 过大，导致模型过拟合。这在数学上被称为高方差（High Variance）。
- 衡量指标：有效样本量（Effective Sample Size）：权重和的平方 / 权重的平方和，
$$m^* = \frac{||\alpha||_1^2}{||\alpha||_2^2},\ \alpha=(\alpha _0, \alpha _1, ..., \alpha _m)$$
$\quad$ $m^*$ 最大为 m（所有权重相等时），最小为1（只有一个权重非0时），越小表示权重越集中。 

- 修正方案：**权重截断（Clipped weights）**
    - 引入常数 $C$ 进行截断：$\overline{\alpha}_i = \min(\alpha_i, C)$
    - 可以降低方差（有效样本变多），但会使对协变量偏移的修正效果减弱 —— 偏差-方差权衡（Bias-Variance Tradeoff）
